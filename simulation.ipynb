```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Data for Smart Logistics
data = []
base_lat, base_lon = 14.5995, 120.9842

for _ in range(100):
    record = {
        "timestamp": datetime.now() - timedelta(minutes=np.random.randint(0, 1440)),
        "rfid_tag": f"RFID-{np.random.randint(1000, 9999)}",
        "latitude": round(base_lat + np.random.uniform(-0.01, 0.01), 6),
        "longitude": round(base_lon + np.random.uniform(-0.01, 0.01), 6),
        "temp_c": round(np.random.uniform(2.0, 8.0), 2),
        "status": np.random.choice(["In Transit", "Delivered"])
    }
    data.append(record)

df = pd.DataFrame(data)
df.to_csv("logistics_data.csv", index=False)
print("Success! File created.")
df.head()
```
